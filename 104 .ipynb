```python
def get_list_page(jobname,number):
    for page in range(1,number):
        url = job_url.format(jobname)
        resp = requests.get(url + '&page=' + str(number))
        job_data = resp.json()['data']
        jobs.extend(job_data)                               
        time.sleep(1.5) 
```

```python
def gettotal_page(jobname):
    url = job_url.format(jobname)
    total = int(re.findall('"TOTALPAGE":"(\d+)"',requests.get(url).text)[0])
    return total
```

```python
for row in jobs:
    if 'JOBSKILL_ALL_DESC' not in row:
        row['JOBSKILL_ALL_DESC'] =''
    if 'PCSKILL_ALL_DESC' not in row:
        row['PCSKILL_ALL_DESC'] = ''
for row in jobs:        
        joblist['Title'].append(row['JOB'])
        joblist['Company'].append(row['NAME'])
        joblist['Area'].append(row['JOB_ADDR_NO_DESCRIPT'])
        joblist['Skill'].append(row['PCSKILL_ALL_DESC'])
        joblist['Date'].append(row['APPEAR_DATE'])
    
```
